```python
import os
import gdown
import zipfile

# --- Configuration ---
# Replace this with your Google Drive file ID.
file_id = '1NgWX5YfEKbloAKX9l8kUVJFpWFlUO8UT'
# The URL format for gdown is as follows:
url = f'https://drive.google.com/uc?id={file_id}'

# Name for the downloaded zip file.
zip_filename = 'bdd_dataset.zip'
# Directory where the dataset will be extracted.
extract_to = './bdd_dataset'

# Create the extraction directory if it doesn't exist.
os.makedirs(extract_to, exist_ok=True)

# --- Download the Zip File ---
print("Downloading the dataset from Google Drive...")
gdown.download(url, zip_filename, quiet=False)

# --- Unzip the Downloaded File ---
print("Extracting the dataset...")
with zipfile.ZipFile(zip_filename, 'r') as zip_ref:
    zip_ref.extractall(extract_to)

print("Download and extraction complete!")
```

```python
import subprocess
import sys

# List of required packages
REQUIRED_PACKAGES = [
    "torch",
    "torchvision",
    "matplotlib",
    "tensorboard",
    "tqdm",
    "tabulate",
    "numpy",
    "opencv-python",
    "webbrowser",
    "albumentations"
]

def check_and_install(package):
    try:
        __import__(package)
        print(f"✅ {package} is already installed.")
    except ImportError:
        print(f"❌ {package} is not installed. Installing now...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", package])
        print(f"✅ {package} successfully installed.")

# Checking and installing required packages
if __name__ == "__main__":
    print("\n🔍 Checking for required packages...\n")
    for pkg in REQUIRED_PACKAGES:
        check_and_install(pkg)
    
    print("\n🚀 All required packages are installed and ready to use!")
```
